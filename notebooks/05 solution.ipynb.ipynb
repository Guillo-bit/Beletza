# Conclusiones Finales - Línea Joven (18-25 años)

## 1. Estado de la Línea Joven

- **Ventas totales y unidades vendidas**:  
  - Se detectan caídas intermitentes en Junio, Agosto y Diciembre.  
  - Febrero presenta el pico más alto en ventas y ROI marketing.  

- **Visitas web**:  
  - Constantes entre 17 y 27 visitas promedio por mes.  
  - La caída en ventas no se explica por falta de tráfico, sino por baja conversión.  

- **Tasa de conversión web y add-to-cart**:  
  - Tasa de conversión muy baja en varios meses (0 en Junio, Agosto y Abril).  
  - Engagement con “Add to Cart” también bajo (<25%), mostrando que los visitantes no completan la intención de compra.  

- **ROI de marketing**:  
  - Muy variable, con máximo en Febrero (35.77) y mínimo en Abril (9.57).  
  - Indica que parte de la inversión digital no genera conversiones efectivas.  

- **Ingresos totales**:  
  - Correlacionan con ventas y tasa de conversión.  
  - Máximo en Febrero, mínimo en Diciembre.  

**Insight clave:** La caída en ventas no está relacionada con la inversión ni con el tráfico web, sino con **problemas de conversión y engagement**.

---

## 2. Hipótesis basadas en datos

1. **Experiencia web**:  
   - Proceso de compra complejo o desincentivante.  
   - Carrito abandonado y baja conversión de visitas en ventas.  

2. **Marketing digital**:  
   - Mensajes, canales o promociones no alineadas con el público joven.  
   - ROI bajo indica inversión mal dirigida o creatividades poco efectivas.  

3. **Oferta y percepción de valor**:  
   - Aunque el precio promedio varía, no parece explicar las caídas.  
   - Posible desconexión entre la propuesta de la línea y expectativas de los jóvenes.

## 4. Recomendaciones estratégicas

**Optimización rápida y de bajo costo:**

1. **Experiencia web y conversión**  
   - Simplificar flujo de compra.  
   - Recordatorios y seguimiento de carritos abandonados.  
   - Mejora de descripciones y reseñas relevantes para jóvenes.  

2. **Marketing segmentado**  
   - Micro-campañas en redes sociales y con influencers.  
   - Promociones limitadas para incentivar urgencia.  
   - Medición de ROI por campaña para ajustes rápidos.  

3. **Engagement y retención**  
   - Incentivos por acciones web (clics, add-to-cart, compartir).  
   - Encuestas rápidas para ajustar la oferta y la comunicación.  

---

## 5. Impacto esperado

- **Incremento de ventas**: 10-20% mensual si la conversión aumenta del 5-15% al 15-25%.  
- **Tasa de conversión web**: subir al 15% promedio.  
- **ROI marketing**: mejorar ROI general >20.  
- **Engagement**: aumento de add-to-cart y clics medibles con micro-campañas.
